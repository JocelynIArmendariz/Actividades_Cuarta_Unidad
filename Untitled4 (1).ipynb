```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

#Generar aleatoriamente las edades de los 500 estudiantes
media = 70
desviacion_estandar = 4
cantidad_estudiantes = 500

edades_estudiantes = np.random.normal(media, desviacion_estandar, cantidad_estudiantes)

# Realizar un histograma de las edades
plt.figure(figsize=(10, 6))
plt.hist(edades_estudiantes, bins=30, color='skyblue', edgecolor='blue', alpha=0.7)
plt.title('Histograma de Pesos de los Estudiantes')
plt.xlabel('Peso (kg)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Graficar la funcin de densidad de probabilidad
plt.figure(figsize=(10, 6))
plt.plot(sorted(edades_estudiantes), norm.pdf(sorted(edades_estudiantes), media, desviacion_estandar), color='pink')
plt.title('Funcin de Densidad de Probabilidad')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```
